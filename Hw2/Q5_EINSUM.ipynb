```
import torch

# Example
a = torch.arange(6).reshape(2, 3)
torch.einsum('ij->ji', [a])
```

```
#Q5.1
q1a = torch.randn(3,2,5)
q1b = torch.randn(3)
torch.einsum('ijk,i->jk',[q1a,q1b])
```

```
#Q5.2
q2 = torch.randn(3,2,5,3)
torch.einsum('ijkl->ik',[q2])
```

```
#Q5.3
q3 = torch.randn(3,2,5,3)
torch.einsum('ijkl->ki',[q3])
```

```
#Q5.4
q4 = torch.randn(3,2,5)
torch.einsum('ijk,ijk->i',[q4,q4])
```

```
#Q5.5
q5a = torch.arange(6).reshape(2, 3)
q5g = torch.arange(6).reshape(2, 3)
q5gt = q5g.t()
q5b = torch.arange(6).reshape(2, 3)
torch.einsum('de,ef,fl->dl',[q5a,q5gt,q5b])
```
